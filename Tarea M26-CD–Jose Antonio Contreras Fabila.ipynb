```python
import os
import csv
import numpy as np
import pandas as pd
from statistics import mean, median, mode, stdev
import plotly as px
os.getcwd() # Directorio actual
os.chdir('/users/yoga7i/downloads') # Cambiar directorio de trabajo
file = open('Height_of_Male_and_Female_by_Country_2022.csv', 'r', encoding='latin1') # Abrir archivo csv
csvreader = csv.reader(file) # Leer archivo csv
rows = [] # Crear lista vacía
for row in csvreader: # Iterar sobre filas
    rows.append(row) # Agregar fila a la lista
print(rows[:50]) # Mostrar las primeras cinccuenta filas




```

```python
import numpy as np
height_male_cm = np.array(rows) # Generamos el objeto height con todas las filas de la lista
# height_male_cm[0:49,2] # Seleccionamos las filas de la 0 a la 49 y la columna 2
height_male_cm = height_male_cm[1:] # Excluir el encabezado
height_male_cm_int = height_male_cm[:,2].astype(float) # Convertir a tipo float
height_male_cm_int
mean_height = mean(height_male_cm_int) # Calcular la media
min_height_male_cm = min(height_male_cm_int) # Calcular el valor mínimo
max_height_male_cm = max(height_male_cm_int) # Calcular el valor mínimo
print(f"La estatura media de los hombres es: {mean_height:.2f} cms")
print(f"La estatura mínima de los hombres es: {min_height_male_cm} cms")
print(f"La estatura máxima de los hombres es: {max_height_male_cm} cms")
```

```python
import numpy as np
height_female_cm = np.array(rows) # Generamos el objeto height con todas las filas de la lista
# height_female_cm[0:49,2] # Seleccionamos las filas de la 0 a la 49 y la columna 2
height_female_cm = height_female_cm[1:] # Excluir el encabezado
height_female_cm_int = height_female_cm[:,3].astype(float) # Convertir a tipo float
height_female_cm_int
mean_height = mean(height_female_cm_int) # Calcular la media
min_height_female_cm = min(height_female_cm_int) # Calcular el valor mínimo
max_height_female_cm = max(height_female_cm_int) # Calcular el valor mínimo
print(f"La estatura media de las mujeres es: {mean_height:.2f} cms")
print(f"La estatura mínima de las mujeres es: {min_height_female_cm} cms")
print(f"La estatura máxima de las mujeres es: {max_height_female_cm} cms")
```

```python
import pandas as pd
import plotly.express as px

# Cargar el archivo CSV
df = pd.read_csv("Height_of_Male_and_Female_by_Country_2022.csv")

# Filtrar solo México
mexico = df[df["Country Name"] == "Mexico"]

# Preparar datos en formato largo para comparación
df_melted = mexico.melt(
    value_vars=["Male Height in Cm", "Female Height in Cm"],
    var_name="Sexo",
    value_name="Altura (cm)"
)

# Renombrar categorías para mejor presentación
df_melted["Sexo"] = df_melted["Sexo"].replace({
    "Male Height in Cm": "Hombres",
    "Female Height in Cm": "Mujeres"
})

# Crear gráfico de barras
fig = px.bar(
    df_melted,
    x="Sexo",
    y="Altura (cm)",
    title="Altura promedio en México por sexo (2022)",
    text="Altura (cm)",
    color="Sexo"
)

fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig.update_layout(yaxis_title="Altura en cm", xaxis_title="Sexo")
fig.show()
```
